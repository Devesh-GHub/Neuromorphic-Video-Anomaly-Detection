```python
import pandas as pd

data = [
    {"Experiment": "MSE", "AUC": 0.82},
    {"Experiment": "MAE", "AUC": 0.79},
    {"Experiment": "MSE+MAE", "AUC": 0.83},
    {"Experiment": "LR=1e-4", "AUC": 0.80},
    {"Experiment": "Latent=64", "AUC": 0.81},
]

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Experiment"], df["AUC"])
plt.ylabel("AUC-ROC")
plt.title("Ablation Study Results")
plt.xticks(rotation=45)
plt.show()
```
