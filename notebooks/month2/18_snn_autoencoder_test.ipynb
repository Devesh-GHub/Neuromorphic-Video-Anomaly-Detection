```python
import numpy as np
import os
import sys
import importlib

sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), "../..")))

from models import snn_autoencoder
importlib.reload(snn_autoencoder)
from models.snn_autoencoder import SNNAutoencoder
```

```python
autoencoder = SNNAutoencoder(
    input_shape=(2, 32, 32),
    hidden_channels=(16, 8, 4),
    simulation_time=50
)
autoencoder.initialize_weights()
```

```python
C, H, W, T = 2, 32, 32, 50
input_spikes = np.zeros((C, H, W, T), dtype=np.float32)

# Random sparse spikes
input_spikes[np.random.rand(C, H, W, T) < 0.01] = 1
```

```python
output = autoencoder.forward(input_spikes)

print(output['spike_counts'])
```

```python
print("Input spikes:", input_spikes.sum())
print("Output spikes:", output['output_spikes'].sum())
print("Bottleneck activity:", output['bottleneck_activity'].sum())
```
